### System Pipeline Overview

The system follows a four-stage AI reasoning pipeline:

1. **Data Understanding:** The input (X) includes key cloud requirements such as cost, SLA, and region.
2. **Inference:** Gemini interprets the requirements and maps them to relevant providers and configurations.
3. **Reasoning:** The model compares multiple providers based on trade-offs (cost, performance, scalability).
4. **Output Generation:** The model outputs a recommendation (y) and a human-readable explanation.

This process demonstrates an intelligent decision support loop using AI reasoning and prompt engineering.
